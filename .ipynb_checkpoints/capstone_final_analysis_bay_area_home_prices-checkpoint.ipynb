## 1. Business Problem & Research Question

San Jose sits at the center of Silicon Valley and has experienced extreme housing price growth over the past two decades. Buyers, sellers, and policymakers all care about:

- How expensive different ZIP codes are (and how they are segmented: entry-level vs luxury).
- How those prices evolve over time relative to the broader Bay Area.
- To what extent we can **predict near-term home values** from local and macro-economic indicators.

**Research Question**

> How can machine learning models accurately predict housing prices in San Jose while uncovering distinct market segments (e.g., entry-level, mid-market, luxury ZIPs) that behave differently under changing economic and local conditions?

We tackle this in two parts:

1. **Segmentation / EDA:**  
   - Compare San Jose home values vs the rest of the Bay Area.  
   - Cluster San Jose ZIP codes (and Bay Area ZIPs) into market segments.

2. **Modeling:**  
   - Build baseline regression models to predict ZIP-level ZHVI.  
   - Compare models **with vs without** lagged home values (`zhvi_lag1`).  
   - Evaluate performance with RMSE and R², and interpret feature importance.

## 2. Imports & configuration

```python
# 2. Imports & configuration

from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error

# Plotting style
sns.set(style="whitegrid", context="notebook")
plt.rcParams["figure.dpi"] = 110

RANDOM_STATE = 42

PROC = Path("data/processed")
assert PROC.exists(), "Expected data/processed directory to exist."
```

### Data Sources

This notebook uses the cleaned and merged panel that was prepared in Module 20:

- `data/processed/zillow_zhvi_zip_long.csv` – long-form Zillow Home Value Index (ZHVI) by ZIP/month.
- `data/processed/redfin_zip_monthly.csv` – Redfin ZIP-level monthly metrics  
  (`median_sale_price`, `median_list_price`, `inventory`, `homes_sold`, etc.).
- `data/processed/fred_monthly.csv` – FRED macroeconomic time series  
  (`FEDFUNDS`, `CPIAUCSL`, `UNRATE`).
- `data/processed/panel_zip_monthly.csv` – final merged panel (Bay Area ZIP × month) with:
  - ZHVI,
  - Redfin metrics,
  - lagged ZHVI & 1-month % change,
  - sales/inventory ratio,
  - macro variables for each month.

In this notebook we **consume** `panel_zip_monthly.csv` and focus on:

- San Jose ZIP codes for **modeling**, and  
- Bay Area ZIP codes for comparison and clustering.


#### Load Panel & Basic Cleaning

```python
# 3. Load merged panel & basic cleaning

panel_path = PROC / "panel_zip_monthly.csv"
panel = pd.read_csv(panel_path, parse_dates=["date"])

print("Panel shape:", panel.shape)
print("Columns:", panel.columns.tolist())

# Ensure ZIP is treated as string with leading zeros
panel["zip"] = panel["zip"].astype(str).str.zfill(5)

# Identify San Jose ZIPs
panel["city_lower"] = panel["city"].str.lower()
panel["is_san_jose"] = panel["city_lower"].eq("san jose")

# Drop columns we know we don't want in modeling (e.g., nearly fully missing)
if "price_drops" in panel.columns:
    print("Dropping 'price_drops' (nearly 100% missing in this subset).")
    panel = panel.drop(columns=["price_drops"])

# Quick sanity check on date range
print("Date range:", panel["date"].min(), "→", panel["date"].max())

# Keep only rows with reasonable ZHVI values to remove extreme outliers / bad joins
panel = panel[(panel["zhvi"] > 50_000) & (panel["zhvi"] < 5_000_000)]
print("After ZHVI filter, panel shape:", panel.shape)
```

## 4. Market Segmentation – Clustering ZIPs

To address the **segmentation** part of the problem statement, we:

1. Aggregate ZIP-level data across time.
2. Build a ZIP summary with:
   - Median ZHVI,
   - Average monthly growth,
   - Volatility of month-to-month change,
   - Average inventory and sales/inventory ratio.
3. Apply **K-Means clustering** to identify market segments:
   - Entry-level
   - Mid-market
   - High-price / Luxury

We start with **San Jose ZIPs**, then repeat the analysis Bay Area-wide.


#### San Jose ZIP Clustering

```python
# 4.1 San Jose ZIP clustering

# Construct per-ZIP summary for San Jose
sj_zip_summary = (
    panel_sj
    .groupby("zip")
    .agg(
        median_zhvi=("zhvi", "median"),
        avg_zhvi=("zhvi", "mean"),
        volatility=("zhvi_pct_change_1m", "std"),
        mean_sales_inv=("sales_inventory_ratio", "mean"),
        avg_inventory=("inventory", "mean"),
        avg_homes_sold=("homes_sold", "mean"),
        avg_growth=("zhvi_pct_change_1m", "mean"),
    )
    .dropna()
    .reset_index()
)

print("San Jose ZIP summary shape:", sj_zip_summary.shape)
sj_zip_summary.head()
```

#### KMeans on San Jose ZIPs + Labels

```python
# Features for clustering
cluster_features_sj = [
    "avg_zhvi",
    "avg_growth",
    "volatility",
    "mean_sales_inv",
    "avg_inventory",
    "avg_homes_sold",
]

X_sj = sj_zip_summary[cluster_features_sj].values

scaler_sj = StandardScaler()
X_sj_scaled = scaler_sj.fit_transform(X_sj)

# KMeans clustering (3 segments)
k_sj = 3
kmeans_sj = KMeans(
    n_clusters=k_sj,
    random_state=RANDOM_STATE,
    n_init=10
)
sj_clusters = kmeans_sj.fit_predict(X_sj_scaled)

sj_zip_summary["cluster"] = sj_clusters

# Order clusters by median price and assign human-readable names
cluster_order = (
    sj_zip_summary.groupby("cluster")["avg_zhvi"]
    .median()
    .sort_values()
    .index
    .tolist()
)

segment_names_sj = {
    cluster_order[0]: "Entry-level",
    cluster_order[1]: "Mid-market",
    cluster_order[2]: "High-price / Luxury",
}

sj_zip_summary["segment"] = sj_zip_summary["cluster"].map(segment_names_sj)

sj_zip_summary.sort_values("avg_zhvi").head()
```

#### Show Segments and ZIP Lists (San Jose)

```python
# Inspect San Jose segments and list ZIPs

print("San Jose ZIP segments:")
print(
    sj_zip_summary.groupby("segment")["zip"]
    .count()
    .rename("num_zips")
)

for seg in ["Entry-level", "Mid-market", "High-price / Luxury"]:
    zips_in_seg = (
        sj_zip_summary[sj_zip_summary["segment"] == seg]
        .sort_values("avg_zhvi")["zip"]
        .tolist()
    )
    print(f"\n{seg} ({len(zips_in_seg)} ZIPs):")
    print(", ".join(zips_in_seg))
```

```python
# 4.2 Visualizing San Jose segments in 2D

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=sj_zip_summary,
    x="avg_zhvi",
    y="mean_sales_inv",
    hue="segment",
    s=80
)
plt.title("San Jose ZIP segments – avg ZHVI vs sales/inventory ratio")
plt.xlabel("Average ZHVI")
plt.ylabel("Average sales/inventory ratio")
plt.legend(title="Segment")
plt.tight_layout()
plt.show()
```

**Interpretation – San Jose segments**

- **Entry-level** ZIPs:
  - Lower average ZHVI,
  - Often higher sales/inventory ratio (homes move relatively quickly),
  - These areas behave more like “starter” neighborhoods.

- **Mid-market** ZIPs:
  - Mid-range ZHVI,
  - Moderate sales/inventory ratio,
  - Represent “move-up” neighborhoods with a balance of price and activity.

- **High-price / Luxury** ZIPs:
  - Highest ZHVI by a clear margin,
  - Sometimes slightly lower sales/inventory ratio and smaller inventory,
  - Represent premium neighborhoods with constrained supply and strong demand.

These segments address the **“distinct market segments”** part of the problem statement.


#### Bay Area-wide ZIP Clustering

```python
# 4.3 Bay Area-wide ZIP clustering

zip_summary_bay = (
    panel
    .groupby("zip")
    .agg(
        median_zhvi=("zhvi", "median"),
        avg_zhvi=("zhvi", "mean"),
        volatility=("zhvi_pct_change_1m", "std"),
        mean_sales_inv=("sales_inventory_ratio", "mean"),
        avg_inventory=("inventory", "mean"),
        avg_homes_sold=("homes_sold", "mean"),
        avg_growth=("zhvi_pct_change_1m", "mean"),
    )
    .dropna()
    .reset_index()
)

cluster_features_bay = [
    "avg_zhvi",
    "avg_growth",
    "volatility",
    "mean_sales_inv",
    "avg_inventory",
    "avg_homes_sold",
]

X_bay = zip_summary_bay[cluster_features_bay].values
scaler_bay = StandardScaler()
X_bay_scaled = scaler_bay.fit_transform(X_bay)

k_bay = 3
kmeans_bay = KMeans(
    n_clusters=k_bay,
    random_state=RANDOM_STATE,
    n_init=10
)
bay_clusters = kmeans_bay.fit_predict(X_bay_scaled)

zip_summary_bay["cluster"] = bay_clusters

# Label clusters by price level for Bay Area as well
cluster_order_bay = (
    zip_summary_bay.groupby("cluster")["avg_zhvi"]
    .median()
    .sort_values()
    .index
    .tolist()
)

segment_names_bay = {
    cluster_order_bay[0]: "Entry-level",
    cluster_order_bay[1]: "Mid-market",
    cluster_order_bay[2]: "High-price / Luxury",
}

zip_summary_bay["segment"] = zip_summary_bay["cluster"].map(segment_names_bay)

zip_summary_bay.head()
```

#### Bay Area Scatter Plot with San Jose Highlighted

```python
# 4.4 Plot Bay Area segments, highlighting San Jose ZIPs

# Add a flag for San Jose
zip_summary_bay = zip_summary_bay.merge(
    panel[["zip", "city_lower"]].drop_duplicates(),
    on="zip",
    how="left"
)
zip_summary_bay["is_san_jose"] = zip_summary_bay["city_lower"].eq("san jose")

plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=zip_summary_bay,
    x="avg_zhvi",
    y="mean_sales_inv",
    hue="segment",
    style="is_san_jose",
    s=60,
    alpha=0.8
)
plt.title("Bay Area ZIP segments – San Jose highlighted")
plt.xlabel("Average ZHVI")
plt.ylabel("Average sales/inventory ratio")
plt.legend(title="Segment / San Jose")
plt.tight_layout()
plt.show()
```

**Interpretation – Bay Area vs San Jose**

- The Bay Area as a whole also breaks into **3 broad tiers**:
  - Entry-level,
  - Mid-market,
  - High-price / Luxury.
- **San Jose ZIPs** tend to cluster:
  - Toward the **mid-market and high-price** segments,
  - With relatively few ZIPs in the truly entry-level range by Bay Area standards.
- This confirms that San Jose is:
  - A relatively expensive sub-market within an already expensive region,
  - With meaningful internal segmentation between more affordable and luxury ZIPs.


## 5. Modeling – Predicting ZHVI at ZIP × Month Level (San Jose)

We now build regression models to predict **ZHVI for San Jose ZIPs** at the ZIP × month level.

We construct two feature sets:

1. **With lag (`zhvi_lag1`) – “full” model**  
   - Uses `zhvi_lag1` plus contemporaneous Zillow, Redfin, and macro features.

2. **Without lag – “contemporaneous only” model**  
   - Uses only Redfin and macro features (no lagged ZHVI),
   - This tests how far we can go without directly relying on yesterday’s price.

We compare four models:

- Linear Regression (**with lag**)
- Random Forest Regression (**with lag**)
- Linear Regression (**without lag**)
- Random Forest Regression (**without lag**)

All models share the **same train/test split** for fair comparison.


#### Prepare Modeling Dataset (San Jose)

```python
# 5.1 Prepare modeling dataset for San Jose

model_df = panel_sj.copy()

core_features = [
    "zhvi_lag1",
    "zhvi_pct_change_1m",
    "median_sale_price",
    "median_list_price",
    "inventory",
    "homes_sold",
    "sales_inventory_ratio",
    "FEDFUNDS",
    "CPIAUCSL",
    "UNRATE",
]

target_col = "zhvi"

# Ensure no weird infinities
model_df[core_features + [target_col]] = model_df[core_features + [target_col]].replace(
    [np.inf, -np.inf], np.nan
)

# Drop rows with missing target
model_df = model_df.dropna(subset=[target_col])

# For the feature set WITH lag, allow SimpleImputer to handle NaNs
print("Modeling dataset shape (San Jose):", model_df.shape)

model_df[core_features + [target_col]].isna().mean().sort_values()
```

#### Train/Test Split Indices

```python
# 5.2 Train/test split indices (reused across all models)

# We'll define X_full using lag features; y is always ZHVI
X_full = model_df[core_features].copy()
y = model_df[target_col].copy()

X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_full,
    y,
    test_size=0.20,
    random_state=RANDOM_STATE
)

train_idx = X_train_full.index
test_idx = X_test_full.index

print("Train size:", len(train_idx), "| Test size:", len(test_idx))
```

#### Linear Regression WITH Lag

```python
# 5.3 Linear Regression – WITH zhvi_lag1

lin_with_lag = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

lin_with_lag.fit(X_train_full, y_train)
y_pred_lin_full = lin_with_lag.predict(X_test_full)

rmse_lin_full = root_mean_squared_error(y_test, y_pred_lin_full)
r2_lin_full = r2_score(y_test, y_pred_lin_full)

print("Linear Regression (WITH lag) – level target")
print(f"RMSE: {rmse_lin_full:,.0f}")
print(f"R²:   {r2_lin_full:.3f}")
```

#### Random Forest WITH Lag

```python
# 5.4 Random Forest – WITH zhvi_lag1

rf_with_lag = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestRegressor(
        n_estimators=200,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        max_depth=None,
    )),
])

rf_with_lag.fit(X_train_full, y_train)
y_pred_rf_full = rf_with_lag.predict(X_test_full)

rmse_rf_full = root_mean_squared_error(y_test, y_pred_rf_full)
r2_rf_full = r2_score(y_test, y_pred_rf_full)

print("Random Forest (WITH lag) – level target")
print(f"RMSE: {rmse_rf_full:,.0f}")
print(f"R²:   {r2_rf_full:.3f}")
```

**Interpretation – models with `zhvi_lag1`**

- Both Linear Regression and Random Forest achieve:
  - **Very high R²** (close to 1.0), and
  - **Low RMSE** relative to the typical level of ZHVI in San Jose.
- This is expected:
  - ZHVI is a **slow-moving time series**,
  - The best predictor of today’s value is **yesterday’s value (`zhvi_lag1`)** plus local price-level features.

However, relying heavily on `zhvi_lag1` can limit the model’s usefulness for **scenario analysis** (e.g., what happens if inventory doubles?).  
To measure the value of non-lag features, we now repeat the analysis **without** `zhvi_lag1`.


#### Build NO-LAG Feature Set & Models

In this subsection we build no-lag models on the **same San Jose modeling dataset as before** (`model_df`, all property types), to make a direct apples-to-apples comparison with the lagged models.


```python
# 5.5 Feature set WITHOUT lag

no_lag_features = [
    "median_sale_price",
    "median_list_price",
    "inventory",
    "homes_sold",
    "sales_inventory_ratio",
    "FEDFUNDS",
    "CPIAUCSL",
    "UNRATE",
]

X_no_lag = model_df.loc[train_idx.union(test_idx), no_lag_features].copy()
y_aligned = model_df.loc[train_idx.union(test_idx), target_col]

# Split using the same index sets
X_train_nl = X_no_lag.loc[train_idx]
X_test_nl = X_no_lag.loc[test_idx]
y_train_nl = y_aligned.loc[train_idx]
y_test_nl = y_aligned.loc[test_idx]

# Linear Regression without lag
lin_no_lag = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

lin_no_lag.fit(X_train_nl, y_train_nl)
y_pred_lin_nl = lin_no_lag.predict(X_test_nl)

rmse_lin_nl = root_mean_squared_error(y_test_nl, y_pred_lin_nl)
r2_lin_nl = r2_score(y_test_nl, y_pred_lin_nl)

print("Linear Regression (NO lag) – level target")
print(f"RMSE: {rmse_lin_nl:,.0f}")
print(f"R²:   {r2_lin_nl:.3f}")

# Random Forest without lag
rf_no_lag = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestRegressor(
        n_estimators=200,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        max_depth=None,
    )),
])

rf_no_lag.fit(X_train_nl, y_train_nl)
y_pred_rf_nl = rf_no_lag.predict(X_test_nl)

rmse_rf_nl = root_mean_squared_error(y_test_nl, y_pred_rf_nl)
r2_rf_nl = r2_score(y_test_nl, y_pred_rf_nl)

print("\nRandom Forest (NO lag) – level target")
print(f"RMSE: {rmse_rf_nl:,.0f}")
print(f"R²:   {r2_rf_nl:.3f}")
```

```python
# 5.6 Model comparison table

results = pd.DataFrame([
    {
        "Model": "Linear (with lag)",
        "Uses_lag": True,
        "RMSE": rmse_lin_full,
        "R2": r2_lin_full,
    },
    {
        "Model": "Random Forest (with lag)",
        "Uses_lag": True,
        "RMSE": rmse_rf_full,
        "R2": r2_rf_full,
    },
    {
        "Model": "Linear (no lag)",
        "Uses_lag": False,
        "RMSE": rmse_lin_nl,
        "R2": r2_lin_nl,
    },
    {
        "Model": "Random Forest (no lag)",
        "Uses_lag": False,
        "RMSE": rmse_rf_nl,
        "R2": r2_rf_nl,
    },
])

results.sort_values(["Uses_lag", "R2"], ascending=[False, False])
```

**Comparison – with vs without `zhvi_lag1`**

- **With lag**:
  - R² ≈ very close to 1.0,
  - Lowest RMSE,
  - Confirms that **historical ZHVI** plus local price metrics explain most of the variance.

- **Without lag**:
  - R² drops to roughly the **0.8–0.85** range (exact values depend on the random split),
  - RMSE increases significantly,
  - Indicates that current Redfin + macro variables alone capture **some**, but not all, of the signal.

**Takeaways**

- For **short-term forecasting** at the ZIP level, including `zhvi_lag1` is highly beneficial.
- For **scenario analysis** and understanding drivers:
  - Models without lag emphasize the role of contemporaneous variables like `median_sale_price`, `inventory`, and macro factors.
- This directly addresses the Module 20 “next step” of comparing performance **with vs without** `zhvi_lag1`.


#### Random Forest Feature Importance (With Lag)

```python
# 5.7 Feature importance – Random Forest WITH lag

# Extract the underlying RF model and its feature importances
rf_model_full = rf_with_lag.named_steps["model"]
importances = rf_model_full.feature_importances_

fi = pd.DataFrame({
    "feature": core_features,
    "importance": importances
}).sort_values("importance", ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(data=fi, x="importance", y="feature")
plt.title("Random Forest feature importance (WITH lag)")
plt.tight_layout()
plt.show()

fi
```

**Feature importance – WITH lag**

- `zhvi_lag1` is typically the **dominant feature**, confirming that:
  - Past home values are the strongest predictor of current values.
- Price-level metrics (`median_sale_price`, `median_list_price`) also carry substantial importance.
- Macro variables and inventory-related metrics play a **secondary** but non-trivial role.

This aligns with domain intuition:
- Local price history and current listing prices move together,
- Macro conditions (rates, unemployment) shape the overall trend but less of the cross-ZIP variation.


In Sections 5.8–5.10, we refine the no-lag analysis by:

- Restricting to **`property_type == "All Residential"`** (a more homogeneous subset), and
- Training **regularized linear models (Lasso, Ridge)** and a **tuned Random Forest**.

This isolates the effect of contemporaneous features within a single property type and gives a cleaner estimate of how well we can predict ZHVI without using `zhvi_lag1`.


```python
# 5.8 San Jose — no-lag feature matrix for regularized & tree models


# Filter to San Jose + All Residential + sensible price range
sj_df = (
    panel[
        (panel["city"] == "San Jose") &
        (panel["property_type"] == "All Residential") &
        (panel["zhvi"].between(50_000, 5_000_000))
    ]
    .copy()
    .sort_values(["zip", "date"])
)

# 3) Define NO-LAG feature set (note: *no* zhvi_lag1, *no* price_drops)
features_no_lag = [
    "median_sale_price",
    "median_list_price",
    "inventory",
    "homes_sold",
    "sales_inventory_ratio",
    "FEDFUNDS",
    "CPIAUCSL",
    "UNRATE",
]

# Keep rows that have all features + target present
sj_model_nl = sj_df.dropna(subset=features_no_lag + ["zhvi"]).copy()

X_nl = sj_model_nl[features_no_lag]
y_nl = sj_model_nl["zhvi"]

from sklearn.model_selection import train_test_split

X_nl_train, X_nl_test, y_nl_train, y_nl_test = train_test_split(
    X_nl, y_nl, test_size=0.20, random_state=42
)

print("San Jose no-lag sample size:", len(sj_model_nl))
print("Train shape:", X_nl_train.shape, " | Test shape:", X_nl_test.shape)
print("Features used (no lag):", features_no_lag)
```

#### 5.9 Regularized linear models (Lasso & Ridge, no-lag features)

To reduce overfitting and better understand the relative contribution of contemporaneous features, we train:
- **LassoCV** (L1-regularized regression), and
- **RidgeCV** (L2-regularized regression)

using only the no-lag feature set. This complements the plain Linear Regression and allows us to see whether regularization improves out-of-sample performance.


```python
# 5.9 Lasso and Ridge regression (no-lag features)

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error

alphas = np.logspace(-3, 3, 20)

# ----- Lasso -----
lasso_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LassoCV(alphas=alphas, cv=5, random_state=42))
])

lasso_pipe.fit(X_nl_train, y_nl_train)
y_pred_lasso = lasso_pipe.predict(X_nl_test)

rmse_lasso = root_mean_squared_error(y_nl_test, y_pred_lasso)
r2_lasso   = r2_score(y_nl_test, y_pred_lasso)

print("Lasso (San Jose, no-lag features)")
print("Best alpha:", lasso_pipe.named_steps["model"].alpha_)
print(f"RMSE: {rmse_lasso:,.0f}")
print(f"R²:   {r2_lasso:0.3f}")

# ----- Ridge -----
ridge_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", RidgeCV(alphas=alphas, cv=5))
])

ridge_pipe.fit(X_nl_train, y_nl_train)
y_pred_ridge = ridge_pipe.predict(X_nl_test)

rmse_ridge = root_mean_squared_error(y_nl_test, y_pred_ridge)
r2_ridge   = r2_score(y_nl_test, y_pred_ridge)

print("\nRidge (San Jose, no-lag features)")
print("Best alpha:", ridge_pipe.named_steps["model"].alpha_)
print(f"RMSE: {rmse_ridge:,.0f}")
print(f"R²:   {r2_ridge:0.3f}")
```

#### 5.10 Random Forest with GridSearchCV (no-lag features)

We then tune a Random Forest regressor on the same no-lag feature set using GridSearchCV, and inspect feature importances to see which variables drive cross-sectional price differences when past ZHVI is excluded.


```python
# 5.10 Random Forest feature importance (no-lag features)

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns

# Base RF model
rf_nl = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1,
)

rf_param_grid_nl = {
    "max_depth": [6, 10, 14],
    "min_samples_leaf": [1, 3, 5],
}

rf_gs_nl = GridSearchCV(
    rf_nl,
    param_grid=rf_param_grid_nl,
    cv=3,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
)

rf_gs_nl.fit(X_nl_train, y_nl_train)

best_rf_nl = rf_gs_nl.best_estimator_
y_pred_rf_nl = best_rf_nl.predict(X_nl_test)

rmse_rf_nl_tuned = root_mean_squared_error(y_nl_test, y_pred_rf_nl)
r2_rf_nl_tuned   = r2_score(y_nl_test, y_pred_rf_nl)

print("Random Forest (San Jose, no-lag features)")
print("Best params:", rf_gs_nl.best_params_)
print(f"RMSE: {rmse_rf_nl_tuned:,.0f}")
print(f"R²:   {r2_rf_nl_tuned:0.3f}")

# --- Feature importance plot ---
importances = best_rf_nl.feature_importances_
fi = (
    pd.DataFrame({"feature": features_no_lag, "importance": importances})
      .sort_values("importance", ascending=True)
)

plt.figure(figsize=(8, 5))
sns.barplot(data=fi, x="importance", y="feature")
plt.title("Random Forest Feature Importance (San Jose, no-lag features)")
plt.xlabel("Relative importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

fi
```

**Interpretation – no-lag models on All Residential**

- The tuned Random Forest achieves **RMSE ≈ \$75k** and **R² ≈ 0.96** on the All Residential subset.  
  - Given typical ZHVI levels around \$1.1–\$1.3M, this is roughly a **6–7% error window**, close to the ±10% accuracy target from the problem statement.
- Feature importance is dominated by:
  - `median_list_price` and `median_sale_price`, followed by
  - CPI (`CPIAUCSL`) and, to a lesser extent, inventory and sales volume.
- This confirms that even **without lagged ZHVI**, a combination of listing prices and macro conditions can explain most of the variation in San Jose home values within a given property type.

# 6. Summary & Recommendations (Business-Focused)

### 6.1 Key Findings

#### **1. San Jose’s housing market has structurally diverged upward from the rest of the Bay Area.**
San Jose home prices followed the broader Bay Area in the early 2010s but have since pushed consistently above regional averages. This reflects long-term demand drivers such as high-income tech employment, limited land availability, and premium school districts across several neighborhoods.

#### **2. The San Jose market is not monolithic—three distinct price-based segments emerge.**
Clustering analysis across ZIP codes reveals three natural groups:

- **Entry-Level Segment** – Lower ZHVI, higher turnover, more sensitive to supply/demand swings.  
- **Mid-Market Segment** – Predictable price appreciation and balanced housing activity.  
- **High-Price / Luxury Segment** – Highest ZHVI, lowest volatility, and minimal sensitivity to month-to-month macro fluctuations.

These segments help explain why certain neighborhoods behave differently under changing market conditions.

#### **3. San Jose ZIPs align closely with Bay Area-wide cluster patterns.**
When clustering at the Bay Area level, San Jose ZIPs mostly fall into mid-market and luxury tiers—reinforcing its role as a premium-priced submarket in Silicon Valley.
  
---

### 6.2 Modeling Insights

#### **1. Models using lagged price (`zhvi_lag1`) achieve extremely high accuracy.**
Including `zhvi_lag1` yields:

- **R² above 0.99**,  
- **Very low RMSE**,  
- Highly stable predictions month-to-month.

This is expected: past home values are the strongest predictor of current values in a stable housing market. These models are excellent for **short-term forecasting**.

#### **2. Models without lag highlight true economic drivers of price.**
Removing `zhvi_lag1` forces the models to rely solely on contemporaneous variables such as:

- Median sale price  
- Median list price  
- Inventory  
- Sales-to-inventory ratio  
- Key macro variables (interest rates, CPI, unemployment)

These models have lower accuracy but offer **far greater interpretability**, making them suitable for:

- Scenario analysis  
- Understanding ZIP-level differences  
- Policy and planning use cases  

Regularized models (Lasso, Ridge) and Random Forests confirm that **price-level metrics** drive most of the explanatory power, with inventory/sales dynamics and macro conditions adding secondary influence.

#### **3. Tuned Random Forest (no-lag) provides the best balance of accuracy and interpretability.**
This model captures non-linear relationships in supply and demand while avoiding overfitting, making it ideal for business users who need insights rather than perfect predictions.
